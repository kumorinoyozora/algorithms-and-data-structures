```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dots = np.array([
    (90, 50),
    (65, 20),
    (65, 75),
    (35, 26),
    (25, 55),
    (26, 45),
    (19, 28),
    (30, 85),
    (25, 90),
    (5, 95),
    (15, 85),
    (10, 66),
    (11, 16),
    (10, 5),
    (5, 10),
    (5, 45),
    (6, 24),
    (10, 40),
    (45, 45),
    (32, 42),
    (30, 20),
    (46, 96),
    (4, 35),
    (38, 10),
    (51, 55)
])

x = dots[:, 0]
y = dots[:, 1]
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlim(0, 100)
ax.set_ylim(0, 100)

ax.axvline(50, 0, 100)
ax.plot((0, 50), (50, 50))
ax.plot((20, 20), (50, 100))
ax.plot((15, 15), (0, 50))
ax.plot((0, 15), (20, 20))
ax.plot((15, 50), (35, 35))
ax.plot((46), (55), marker='*', ms=15, c='r')

ax.set_xticks(range(0, 101, 5))
ax.set_yticks(range(0, 101, 5))

plt.show()
```
